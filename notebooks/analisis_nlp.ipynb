```python
# Importamos librerías necesarias para NLP y conteo de palabras
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
from nltk.stem import SnowballStemmer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
# Tokenizamos un texto en inglés en palabras y oraciones, y mostramos stopwords
text = "NLP is fascinating. Python makes it easy."
words = word_tokenize(text)
sentences = sent_tokenize(text)
stop_words = set(stopwords.words("english"))
print(words)
print(sentences)
print(stop_words)
```

```python
# Tokenizamos un pequeño texto en español y contamos frecuencia de palabras
texto = "Iba Sancho Panza sobre su jumento como un patriarca, con sus alforjas y su bota."
tokens = word_tokenize(texto, language="spanish")
freq = FreqDist(tokens)
print(tokens)
print(freq.most_common(10))
```

```python
# Creamos la lista de palabras comunes en español
stop_words = set(stopwords.words("spanish"))
print(stop_words)
```

```python
# Aplicamos SnowballStemmer para ver la raíz de algunas palabras
stemmer = SnowballStemmer("spanish")
palabras = ["corriendo", "corrió", "corríamos", "corredores"]
for palabra in palabras:
    print(palabra, "→", stemmer.stem(palabra))
```

```python
# Tokenizamos un texto corto y aplicamos stemming a cada palabra
texto = "Los perros estaban corriendo rápidamente por el parque."
tokens = word_tokenize(texto, language="spanish")
stems = [stemmer.stem(t) for t in tokens]
print("Tokens:")
print(tokens)
print("\nStems:")
print(stems)
```

```python
# Eliminamos palabras comunes que no aportan información
stop_words = set(stopwords.words("spanish"))
tokens_limpios = [t for t in tokens if t.lower() not in stop_words]
print(tokens_limpios)
```

```python
# Abrimos el archivo de la novela "cien.txt" desde la carpeta data
with open("../data/cien.txt", "r", encoding="utf-8") as f:
    texto = f.read()
print(texto[600:1000])
```

```python
# Tokenizamos la novela, quitamos stopwords, aplicamos stemming y contamos frecuencia
tokens = word_tokenize(texto, language="spanish")
stop_words = set(stopwords.words("spanish"))
tokens_limpios = [t for t in tokens if t.lower() not in stop_words]
stemmer = SnowballStemmer("spanish")
stems = [stemmer.stem(t) for t in tokens_limpios]
freq = FreqDist(tokens_limpios)
print(freq.most_common(20))
```

```python
# Graficamos las 10 palabras más frecuentes del texto
freq.plot(10)
```
